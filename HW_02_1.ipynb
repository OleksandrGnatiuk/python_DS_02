Прочитайте дані за допомогою методу read_html з таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)"

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

all_tables = pd.read_html('https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8',
                          thousands="",
                          decimal = ",",
                          match = "Севастополь",
                    )
table = pd.DataFrame(all_tables[1])
table
```

Вивести перші рядки таблиці за допомогою методу head:

```python
table.head(3)
```

Визначте кількість рядків та стовпців у датафреймі (атрибут shape):

```python
table.shape
```

Замініть у таблиці значення "—" на значення NaN:

```python
table.replace( "—", 'NaN', inplace=True)
table
```

Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
table.dtypes
```

Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
table = table.loc[~table['2014'].isna()]
table = table.loc[~table['2019'].isna()]
table['2014'] = table['2014'].astype("float")
table['2019'] = table['2019'].astype("float")
table.dtypes
```

Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
missed_data = table.isnull().sum(axis=0)
missed_data
```

Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
table = table.loc[:26]
table
```

Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
avg = table.mean(axis=0, numeric_only=True)
print("Середні значення по стовпцям:\n", avg)
table.fillna(avg, inplace=True)
table
```

Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
avg = table.mean(numeric_only=True)['2019']
print(f"Середній рівень народжуваності в Україні в 2019 році: {avg}")

table[['Регіон', '2019']][table['2019'] > avg]
```

У якому регіоні була найвища народжуваність у 2014 році?

```python
table[["Регіон", "2014"]][table["2014"] == table["2014"].max()]
```

Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
plt.bar(table['Регіон'],table['2019'], color='r')
plt.xlabel('Регіон', fontsize='small', color='midnightblue', )
plt.xticks(rotation=90, fontsize='small')
plt.ylabel('Коеф. народжуваністі', fontsize='small', color='midnightblue')
plt.title('Діаграма народжуваності по регіонам в 2019 році', fontsize=16, color='b')
plt.show()
```
