Прочитайте дані за допомогою методу read_html з таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)"

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96"

all_tables = pd.read_html(url,
                          thousands="",
                          decimal = ",",
                          match = "Коефіцієнт народжуваності у регіонах України у 1950—2014 роках",
                    )
table = pd.DataFrame(all_tables[0])
table
```

Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
avg = table.mean(axis=0, numeric_only=True)
print("Середні значення по стовпцям: ")
print(avg)
# print('Замінюємо відсутні дані середніми значеннями:!')
table = table.fillna(avg)
table
```

Отримайте список регіонів, де рівень народжуваності у 2014 році був вищим за середній по Україні

```python
avg = table.mean(numeric_only=True)['2014']
print(f"Середній рівень народжуваності в Україні в 2014 році: {avg}")

table[['регіон', '2014']][table['2014'] > avg]
```

У якому регіоні була найвища народжуваність у 2014 році?

```python
table[["регіон", "2014"]][table["2014"] == table["2014"].max()]
```

Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
plt.bar(table['регіон'],table['2014'], color='r')
plt.xlabel('регіон', fontsize='small', color='midnightblue', )
plt.xticks(rotation=90, fontsize='small')
plt.ylabel('Коеф. народжуваністі', fontsize='small', color='midnightblue')
plt.title('Діаграма народжуваності по регіонам в 2014 році', fontsize=16, color='b')
plt.show()
```

```python
df = pd.DataFrame({'Регіон': table['регіон'],
                  '1950':table['1950'],
                   '1990':table['1990'],
                   '2014':table['2014'],
                    })

df = df.set_index('Регіон')

ax = df.plot.barh(y=['1950','1990', '2014'])
plt.title("Коефіцієнт народжуваності в Україні в 1950, 1990, 2014 рр.", color='b')
plt.show()
```
